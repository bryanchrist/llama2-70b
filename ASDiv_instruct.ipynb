```python
import pandas as pd
df = pd.read_csv('ASDiv_clean.csv')
df
```

```python
df['grade'].value_counts()
```

```python
data = []
for i in range(0, len(df)):
    grade = df.iloc[i]['grade']
    output = df.iloc[i]['question']
    instruction = f"Write a grade {grade} math word problem."
    data.append({'instruction': instruction, "input": "", "output": "Question: " + output})
```

```python
import json
myjson = json.dumps(data)
```

```python
with open("ASDiv_instruct.json", "w") as outfile:
    outfile.write(myjson)
```

# Combine ASDiv and GSM8K

```python
import jsonlines
import json
file_path = 'train.jsonl'
data = []

with jsonlines.open(file_path) as reader:
    for line in reader:
        data.append(line)
        
        
for item in data:
    item["output"] = "Question: " + item['question']
    item["instruction"] = "Write a grade school math word problem."
    del item['question']
    del item['answer']
    item['input'] = ""
# Write the data to a new JSON file
output_file = "gsm_clean.json"
with open(output_file, 'w') as json_file:
    json.dump(data, json_file)
```

```python
import json

# Read the first JSON file
file1 = 'gsm_clean.json'
with open(file1, 'r') as json_file:
    data1 = json.load(json_file)

# Read the second JSON file
file2 = 'ASDiv_instruct.json'
with open(file2, 'r') as json_file:
    data2 = json.load(json_file)

# Combine the data from both files
combined_data = data1 + data2

# Write the combined data to a new JSON file
output_file = 'ASDiv_GSM8k.json'
with open(output_file, 'w') as json_file:
    json.dump(combined_data, json_file)
```
