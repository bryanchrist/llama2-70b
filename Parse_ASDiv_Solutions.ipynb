```python
import pandas as pd
import numpy as np
import sys
from io import StringIO
```

```python
file_path = 'mammoth_solve_asdiv.txt'
data = []
unsolved = []
with open(file_path, 'r') as file:
    text = file.read()
    solutions = text.split("Question")
    del solutions[0]
    
    for solution in solutions:
        question_number = solution.split(" ")[1]
        code = solution.split("### Response:")[1].strip()
        initial_vars = dict(locals().items())
        old_stdout = sys.stdout
        redirected_output = sys.stdout = StringIO()
        
        if question_number == "question_843" or question_number == "question_890":
            continue

        try:
            exec(code, globals(), locals())
            sys.stdout = old_stdout

            value = redirected_output.getvalue().strip()
            data.append({'Question Number': question_number, "Solution": value})

            # Create a dictionary to store the variables after execution
            vars_after_exec = dict(locals().items())

            # Find the new variables created by exec()
            new_variables = {var_name: var_value for var_name, var_value in vars_after_exec.items() if var_name not in initial_vars}

            # Delete the new local variables
            for var_name in new_variables:
                del locals()[var_name]

        except Exception as e:
            unsolved.append({'Question Number': question_number, "Solution": str(e)})
            #answers.append(str(e))
```

```python
answers = pd.DataFrame.from_dict(data)
answers.to_csv('ASDiv_raw_solutions.csv')
```
