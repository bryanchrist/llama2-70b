# Concat DFs

```python
df = pd.concat([df1, df2, df3])
df = df.drop_duplicates(subset='question')
df = df.sample(frac = 1)
question_numbers = []
for i in range(0, len(df)):
    question_numbers.append(f"question_{i}")
df['question_number'] = question_numbers
df.to_csv("data\llama_QA_questions.csv")
```

# Send Each Question to .txt file for Zooniverse Upload

```python
for i in range(0, len(df)):
    question_number = df.iloc[i]['question_number']
    question = df.iloc[i]['question']
    solution = df.iloc[i]['solution']
    answer = str(df.iloc[i]['answer'])
    file_name = f"{question_number}.txt"
    file_path = "Zooniverse/"+file_name
    text = "Question Number: " + question_number + "\n\nQuestion:\n" + question + "\n\nSolution:\n" + solution + "\n\nAnswer: " + answer
    with open(file_path, "w") as f:  # Open the file in append mode ("a")
        f.write(text)  # Append the newly generated text to the file
```
