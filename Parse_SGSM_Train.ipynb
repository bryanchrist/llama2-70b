```python
import pandas as pd
sgsm = pd.read_csv('data/sgsm_train.csv')
```

```python
sgsm
```

```python
sgsm = sgsm[['question', 'answer']]
```

```python
for i in range(0, len(sgsm)):
    try:
        answer = sgsm.iloc[i]['answer']
        answer = float(answer)
        answer = str(answer)
        sgsm.iloc[i]['answer'] = answer
    except:
        sgsm = sgsm.drop([i])
```

```python
sgsm
```

```python
sgsm.to_csv('data/sgsm_train_eval.csv', index=False)
```
