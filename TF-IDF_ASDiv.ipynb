```python
import pandas as pd

df = pd.read_json('ASDiv_instruct.json')
```

```python
df
```

```python
grades = ['1', '2', '3', '4', '5', '6']
prompt = f"Write a grade {grade} math word problem."
grade1 = []
grade2 = []
grade3 = []
grade4 = []
grade5 = []
grade6 = []
for grade in grades:
    questions = df.query(f"instruction=='Write a grade {grade} math word problem.'")['output']
    if grade == '1':
        for question in questions:
            grade1.append(question)
    if grade == '2':
        for question in questions:
            grade2.append(question)
            
    if grade == '3':
        for question in questions:
            grade3.append(question)
            
    if grade == '4':
        for question in questions:
            grade4.append(question)
            
    if grade == '5':
        for question in questions:
            grade5.append(question)
            
    if grade == '6':
        for question in questions:
            grade6.append(question)

grade1 = "\n".join(grade1)
grade2 = "\n".join(grade2)
grade3 = "\n".join(grade3)
grade4 = "\n".join(grade4)
grade5 = "\n".join(grade5)
grade6 = "\n".join(grade6)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
pd.set_option("max_rows", 600)
```

```python
data = [grade1, grade2, grade3, grade4, grade5, grade6]
#tfidf_vectorizer = TfidfVectorizer(ngram_range = (3,3), stop_words='english')
tfidf_vectorizer = TfidfVectorizer(ngram_range = (3,3))
tfidf_vector = tfidf_vectorizer.fit_transform(data)
```

```python
feature_names = tfidf_vectorizer.get_feature_names_out()
tfidf_df = pd.DataFrame(tfidf_vector.toarray(), columns=feature_names)
tfidf_df
```

```python
tfidf_df = tfidf_df.stack().reset_index()
tfidf_df
```

```python
tfidf_df = tfidf_df.rename(columns={0:'tfidf', 'level_0': 'document','level_1': 'term', 'level_2': 'term'})
tfidf_df.sort_values(by=['document','tfidf'], ascending=[True,False]).groupby(['document']).head(10)
```
